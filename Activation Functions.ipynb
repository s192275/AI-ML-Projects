```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sigmoid(x):
    a=[]
    for i in x:
        a.append(1/(1+math.exp(-i)))
    return a
```

```python
def tanh(x, derivative = False):
    if(derivative == True):
        return (1-(x**2))
    return np.tanh(x)
        
```

```python
def reLu(x):
    b=[]
    for i in x:
        if i<0:
            b.append(0)
        else:
            b.append(i)
    return b        
```

```python
def leaked_reLu(x):
    b = []
    for i in x:
        if i<0:
            b.append(i/10)
        else:
            b.append(i)
    return b        
```

```python
x = np.arange(-3. ,3. ,0.1)
sig = sigmoid(x)
tanh = tanh(x)
reLu = reLu(x)
leaked_reLu = leaked_reLu(x)
swish = sig * x
```

```python
line_1, = plt.plot(x, sig, label = "Sigmoid")
line_2, = plt.plot(x, tanh, label = "Tanh")
line_3, = plt.plot(x, reLu, label = "ReLu")
line_4, = plt.plot(x, leaked_reLu, label = "Leaked_reLu")
line_5, = plt.plot(x, swish, label = "Swish")

plt.legend(handles = [line_1, line_2, line_3, line_4, line_5])
plt.axhline(y = 0 , color = 'k')
plt.axvline(x = 0, color = 'k')
plt.show()
```
