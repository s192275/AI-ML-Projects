```python
import pandas as pd
import xgboost as xgb
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("Datasets/Social_Network_Ads.csv")
```

```python
data.info()
```

```python
data.describe()
```

```python
data
```

```python
data.groupby(['Gender', 'Age', 'EstimatedSalary'])['Purchased'].agg(['mean'])
```

```python
data['Gender'] = [1 if col == 'Male' else 0 for col in data['Gender']]
```

```python
data
```

```python
Y = data['Purchased']
Y
```

```python
data = data.drop(columns = ['User ID', 'Purchased'])
X = data
```

```python
X
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state= 42)
```

```python
print(X_train.shape)
print(Y_train.shape)
```

```python
X_train, X_val, Y_train, Y_val = train_test_split(X_train, Y_train, test_size = 0.2, random_state = 42)
```

```python
model= xgb.XGBClassifier()
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
y_pred2 = model.predict(X_val)

accuracy = accuracy_score(Y_test, y_pred)
accuracy2 = accuracy_score(Y_val, y_pred2)
print("Accuracy on test:", accuracy)
print('Accuracy on validation: ',accuracy2)
```
